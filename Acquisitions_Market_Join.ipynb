```python
import xlrd
import numpy as np
import pandas as pd
import seaborn as sns
%pylab inline
```

```python
df1 = pd.read_csv('acquisitions+.csv')
```

```python
df2 = pd.read_excel('Acquiree_Market.xlsx')
```

```python
df3 = pd.read_excel('Acquirer_Market.xlsx')
```

```python
Merge_acquiree = df1.merge(df2, left_on = 'Acquiree_Name', right_on = 'company_name', how = 'left')
```

```python
Merge_acquirer = Merge_acquiree.merge(df3, left_on = 'Acquirer Name', right_on = 'acquirer_name', how = 'left')
```

```python
Merge_acquirer['Price']= Merge_acquirer['Price Usd']
```

```python
def format_price(num):
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    # add more suffixes if you need them
    return '%.0f%s' % (num, ['', ' K', ' M', ' B'][magnitude])
```

```python
Merge_acquirer['Price Usd'] = Merge_acquirer['Price Usd'].map(lambda x : format_price(x))
```

```python
Merge_acquirer['Price Usd'] = Merge_acquirer['Price Usd'].map(lambda x : x.replace('nan','Not Disclosed'))
```

```python
Merge_acquirer.drop(['company_name','acquirer_name'],axis=1,inplace=True)
```

```python
Merge_acquirer.rename(columns={'company_market': 'acquiree_market'},inplace=True)
```

```python
Merge_acquirer.to_csv('Final_dataset.csv')
```
